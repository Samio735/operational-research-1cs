```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph({'0':['4'],'1': ['2', '3'], '2': ['4', '3'], '3': ['1'],'4':['1'],'5':[],'6':['5'],'7':['6','5']})
nx.draw(G, with_labels=True)
plt.show()
```

```python
def calc_reachables(V):
    isReachable = {}
    isVisited = {}

    def explore(V):
        isVisited[V] = True
        isReachable[V] = True
        for neighbor in G.neighbors(V):
            if isVisited.get(neighbor):
                pass
            else:
                explore(neighbor)
            isReachable[neighbor] = True
    
    explore(V)
    return isReachable
```

```python
calc_reachables('5')
```
